```python
import pandas as pd

# Yahoo Finance에서 제공 금융 데이터 로드
import yfinance as yf

# 데이터 출처 Yahoo Finance
# 금융 데이터 로드
tickers = ['AAPL','MSFT','AVGO']
start = '2018-01-01'
end = '2023-05-31'
df_ticker_info = pd.DataFrame()
for ticker in tickers:
    ticker_yf = yf.Ticker(ticker)
    if df_ticker_info.empty:
        df_ticker_info = ticker_yf.history(start = start, end = end)
        df_ticker_info['ticker']= ticker
    else:
        data_temp = ticker_yf.history(start = start, end = end)
        data_temp['ticker']= ticker
        df_ticker_info = pd.concat([df_ticker_info, data_temp])

df_ticker_info
```

```python
import numpy as np

# pct_change 함수를 이용, 각 날짜별 종가를 이용하여 수익률을 계산
df_ticker_info['currRet'] = df_ticker_info['Close'].pct_change()

# ticker가 바뀔 때는 정확한 수익률을 계산할 수 없으므로 첫행은 NaN 처리
temp_df_ticker_info = df_ticker_info.loc[:, 'currRet'].copy()
for i in range(len(df_ticker_info)-1):
  if df_ticker_info.iloc[i]['ticker'] != df_ticker_info.iloc[i+1]['ticker']:
    temp_df_ticker_info.iloc[i+1] = np.nan

df_ticker_info['currRet'] = temp_df_ticker_info

# 전일, 당일, 후일의 데이터를 활용하여 이벤트 수익률을 계산하여 이벤트에 따른 주가의 변동을 보다 정확하게 측정
# currRet(t-1) + currRet(t) + currRet(t+1) -> 뉴스 데이터 전일, 당일, 후일의 수익
df_ticker_info['eventRet'] = df_ticker_info['currRet'] + df_ticker_info['currRet'].shift(-1) + df_ticker_info['currRet'].shift(1)

# 날짜 (년-월-일) 정리
df_ticker_info.reset_index(inplace=True)
df_ticker_info['date'] = pd.to_datetime(df_ticker_info['Date']).dt.date

df_ticker_info
```

```python
#finbert 분석
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# FinBERT 모델과 토크나이저 로드
model_name = "ProsusAI/finbert"  # FinBERT 모델
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# 모델의 config 정보에서 레이블 순서 확인
labels = model.config.id2label
print(labels)

# 뉴스 헤드라인을 토큰화하고 PyTorch 텐서로 변환하는 함수
def encode_headlines(headlines, tokenizer):
  encoded_batch = tokenizer.batch_encode_plus(headlines, padding=True, return_tensors='pt', truncation=True, max_length=512)
  return encoded_batch

# 감성 분석을 수행하는 함수
def analyze_sentiment(headlines):
  encoded_batch = encode_headlines(headlines, tokenizer)
  with torch.no_grad():
      outputs = model(**encoded_batch)
      predictions = torch.nn.functional.softmax(outputs.logits, dim=-1)
  return predictions


def analyze_sentiment(headlines):
    encoded_batch = encode_headlines(headlines, tokenizer)
    with torch.no_grad():
        outputs = model(**encoded_batch)
        predictions = torch.nn.functional.softmax(outputs.logits, dim=-1)

    # 클래스별 가중치 설정 (finbert: 긍정=1, 부정=-1, 중립=0)
    weights = torch.tensor([1.0, -1.0, 0.0])

    # 가중치를 활용해 각 예측에 대한 종합 감성 점수 계산
    sentiment_scores = torch.matmul(predictions, weights)

    # 감성 점수를 Python float으로 변환
    sentiment_scores = sentiment_scores.item()

    # 가장 확률이 높은 클래스의 인덱스를 얻음 (예측된 감성 클래스)
    predicted_classes = torch.argmax(predictions, dim=1).item()

    return sentiment_scores, predicted_classes

sentences = filtered_data['headline']

i = 0
finbertSentiment = []
finbertScore = []
for sentence in zip(sentences):
  score, pred_label_index = analyze_sentiment(sentence)

  finbertScore.append(score)
  finbertSentiment.append(pred_label_index)

  i += 1
  if i%100 == 0:
    print(f"i : {i}")

# 결과를 filtered_data 데이터프레임의 새로운 컬럼으로 추가
filtered_data['finbertSentiment'] = finbertSentiment
filtered_data['finbertScore'] = finbertScore

filtered_data
```
