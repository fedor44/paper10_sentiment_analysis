```python
import pandas as pd

# Yahoo Finance에서 제공 금융 데이터 로드
import yfinance as yf

# 데이터 출처 Yahoo Finance
# 금융 데이터 로드
tickers = ['AAPL','MSFT','AVGO']
start = '2018-01-01'
end = '2023-05-31'
df_ticker_info = pd.DataFrame()
for ticker in tickers:
    ticker_yf = yf.Ticker(ticker)
    if df_ticker_info.empty:
        df_ticker_info = ticker_yf.history(start = start, end = end)
        df_ticker_info['ticker']= ticker
    else:
        data_temp = ticker_yf.history(start = start, end = end)
        data_temp['ticker']= ticker
        df_ticker_info = pd.concat([df_ticker_info, data_temp])

df_ticker_info
```

```python
import numpy as np

# pct_change 함수를 이용, 각 날짜별 종가를 이용하여 수익률을 계산
df_ticker_info['currRet'] = df_ticker_info['Close'].pct_change()

# ticker가 바뀔 때는 정확한 수익률을 계산할 수 없으므로 첫행은 NaN 처리
temp_df_ticker_info = df_ticker_info.loc[:, 'currRet'].copy()
for i in range(len(df_ticker_info)-1):
  if df_ticker_info.iloc[i]['ticker'] != df_ticker_info.iloc[i+1]['ticker']:
    temp_df_ticker_info.iloc[i+1] = np.nan

df_ticker_info['currRet'] = temp_df_ticker_info

# 전일, 당일, 후일의 데이터를 활용하여 이벤트 수익률을 계산하여 이벤트에 따른 주가의 변동을 보다 정확하게 측정
# currRet(t-1) + currRet(t) + currRet(t+1) -> 뉴스 데이터 전일, 당일, 후일의 수익
df_ticker_info['eventRet'] = df_ticker_info['currRet'] + df_ticker_info['currRet'].shift(-1) + df_ticker_info['currRet'].shift(1)

# 날짜 (년-월-일) 정리
df_ticker_info.reset_index(inplace=True)
df_ticker_info['date'] = pd.to_datetime(df_ticker_info['Date']).dt.date

df_ticker_info
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import re
import nltk
nltk.download('stopwords')

# 'ticker' 컬럼에서 'MSFT','AAPL', 'AVGO'인 행만 필터링
filtered_data = data[data['ticker'].isin(['MSFT', 'AAPL', 'AVGO'])]

# 불용어 제거
stop_words = set(stopwords.words('english'))
filtered_data['headline'] = filtered_data['headline'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))

# 불필요한 문자 제거 (느낌표,물음표, 달러 표시, 숫자, 점(.)은 유지)
filtered_data['headline'] = filtered_data['headline'].apply(lambda x: re.sub(r'[^a-zA-Z!?$0-9.\'\s]', '', x))

filtered_data
```

```python
import numpy as np

filtered_data['bertSentiment'] = np.nan
filtered_data['finbertSentiment'] = np.nan
filtered_data['finetunedFinbertSentiment'] = np.nan
filtered_data['bertScore'] = np.nan
filtered_data['finbertScore'] = np.nan
filtered_data['finetunedFinbertScore'] = np.nan

filtered_data
```

```python
#finbert 분석
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# FinBERT 모델과 토크나이저 로드
model_name = "ProsusAI/finbert"  # FinBERT 모델
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# 모델의 config 정보에서 레이블 순서 확인
labels = model.config.id2label
print(labels)

# 뉴스 헤드라인을 토큰화하고 PyTorch 텐서로 변환하는 함수
def encode_headlines(headlines, tokenizer):
  encoded_batch = tokenizer.batch_encode_plus(headlines, padding=True, return_tensors='pt', truncation=True, max_length=512)
  return encoded_batch

# 감성 분석을 수행하는 함수
def analyze_sentiment(headlines):
    encoded_batch = encode_headlines(headlines, tokenizer)
    with torch.no_grad():
        outputs = model(**encoded_batch)
        predictions = torch.nn.functional.softmax(outputs.logits, dim=-1)

    # 클래스별 가중치 설정 (finbert: 긍정=1, 부정=-1, 중립=0)
    weights = torch.tensor([1.0, -1.0, 0.0])

    # 가중치를 활용해 각 예측에 대한 종합 감성 점수 계산
    sentiment_scores = torch.matmul(predictions, weights)

    # 감성 점수를 Python float으로 변환
    sentiment_scores = sentiment_scores.item()

    # 가장 확률이 높은 클래스의 인덱스를 얻음 (예측된 감성 클래스)
    predicted_classes = torch.argmax(predictions, dim=1).item()

    return sentiment_scores, predicted_classes

sentences = filtered_data['headline']

i = 0
finbertSentiment = []
finbertScore = []
for sentence in zip(sentences):
  score, pred_label_index = analyze_sentiment(sentence)

  finbertScore.append(score)
  finbertSentiment.append(pred_label_index)

  i += 1
  if i%100 == 0:
    print(f"i : {i}")

# 결과를 filtered_data 데이터프레임의 새로운 컬럼으로 추가
filtered_data['finbertSentiment'] = finbertSentiment
filtered_data['finbertScore'] = finbertScore

filtered_data
```

```python
# 특정 컬럼명 변경 (bert -> vader)
filtered_data = filtered_data.rename(columns={'bertSentiment': 'vaderSentiment', 'bertScore': 'vaderScore'})
filtered_data
```

```python
# 감정 분석을 위해 nltk의 vader 모델을 사용하기 위해 패키지 다운로드
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
nltk.download("all")
import pandas as pd

# 감정 분석 수행
sia = SentimentIntensityAnalyzer()
df_lexicon = pd.read_csv('/content/drive/MyDrive/DeepNeuralNetwork/lexiconData.csv')
# 두 열의 값을 더하고 평균 내어 각 어휘의 감정 점수를 계산
df_lexicon['sentiment'] = (df_lexicon['Aff_Score'] + df_lexicon['Neg_Score'])/2
# Item 열을 키(key)로, sentiment 열을 값(value)으로 갖는 딕셔너리로 변환
df_lexicon = dict(zip(df_lexicon.Item, df_lexicon.sentiment))
# 하나의 단어로 구성된 항목만을 선택하여 딕셔너리 업데이트
df_lexicon = {k:v for k,v in df_lexicon.items() if len(k.split(' '))==1}
temp_dict = {}

# value 조건에 value 값 재계산 후 딕셔너리에 추가
for k, v in df_lexicon.items():
  if v > 0:
    temp_dict[k] = v / max(df_lexicon.values()) * 4
  else:
    temp_dict[k] = v / min(df_lexicon.values()) * -4

final_dict = {}
# 딕셔너리에 주식 시장 관련 어휘와 감정 점수 업데이트
final_dict.update(temp_dict)
# 딕셔너리에 sia.lexicon의 어휘와 감정 점수 업데이트
final_dict.update(sia.lexicon)
# 하나의 딕셔너리로 결합된 정보를 sia.lexicon에 할당
# 주식 시장 관련 어휘와 기존의 감정 어휘를 모두 활용하여 감정 점수 계산 가능
sia.lexicon = final_dict
```

```python
sentences = filtered_data['headline']

# 감성 점수 계산
vaderScore = []
def analyze_sentiment(text):
    score = sia.polarity_scores(text)

    # 일반적 관례에 따라 0.05, -0.05 로 셋팅
    # VADER가 개발될 당시부터 많이 사용되었으며, 대부분의 일반적인 케이스와 데이터셋에서 합리적으로 잘 작동하는 것으로 나타남
    # {0: 'positive', 1: 'negative', 2: 'neutral'} -> ProsusAI/finbert 에 맞추기
    compound_score = score['compound']
    vaderScore.append(compound_score)
    if compound_score >= 0.05:
        return 0 #'Positive'
    elif compound_score <= -0.05:
        return 1 #'Negative'
    else:
        return 2 #'Neutral'

vaderSentiment = []
for sentence in sentences:
  prediction = analyze_sentiment(sentence)

  vaderSentiment.append(prediction)

# 결과를 filtered_data 데이터프레임의 새로운 컬럼으로 추가
filtered_data['vaderSentiment'] = vaderSentiment
filtered_data['vaderScore'] = vaderScore

filtered_data
```

```python
# 뉴스 헤드라인을 토큰화하고 PyTorch 텐서로 변환하는 함수
def encode_headlines(headlines, tokenizer):
  encoded_batch = tokenizer.batch_encode_plus(headlines, padding=True, return_tensors='pt', truncation=True, max_length=512)
  return encoded_batch

# 감성 분석을 수행하는 함수
def analyze_sentiment(headlines):
    encoded_batch = encode_headlines(headlines, tokenizer)
    with torch.no_grad():
        outputs = model(**encoded_batch)
        predictions = torch.nn.functional.softmax(outputs.logits, dim=-1)

    # 클래스별 가중치 설정 (finbert: 긍정=1, 부정=-1, 중립=0)
    weights = torch.tensor([1.0, -1.0, 0.0])

    # 가중치를 활용해 각 예측에 대한 종합 감성 점수 계산
    sentiment_scores = torch.matmul(predictions, weights)

    # 감성 점수를 Python float으로 변환
    sentiment_scores = sentiment_scores.item()

    # 가장 확률이 높은 클래스의 인덱스를 얻음 (예측된 감성 클래스)
    predicted_classes = torch.argmax(predictions, dim=1).item()

    return sentiment_scores, predicted_classes

sentences = filtered_data['headline']

i = 0
finetunedFinbertSentiment = []
finetunedFinbertScore = []
for sentence in zip(sentences):
  score, pred_label_index = analyze_sentiment(sentence)

  finetunedFinbertScore.append(score)
  finetunedFinbertSentiment.append(pred_label_index)

  i += 1
  if i%100 == 0:
    print(f"i : {i}")

# 결과를 filtered_data 데이터프레임의 새로운 컬럼으로 추가
filtered_data['finetunedFinbertSentiment'] = finetunedFinbertSentiment
filtered_data['finetunedFinbertScore'] = finetunedFinbertScore

filtered_data
```
